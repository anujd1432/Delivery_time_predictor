### 2. Feature Engineering & Preprocessing
We will encode categorical variables manually using One-Hot Encoding and drop unnecessary columns like `Order_ID`.

```python
cat_cols=['Weather','Traffic_Level','Time_of_Day','Vehicle_Type']
df=pd.get_dummies(df,columns=cat_cols)
```

```python
df.head()
```

```python
df.isnull().sum()
```

### 3. Data Splitting
Splitting the data into Features (X) and Target (y), then into Training and Testing sets.

```python
x=df.drop(columns=['Order_ID','Delivery_Time_min'])   #independent features
y=df['Delivery_Time_min']  #target feature
```

### 4. Model Training & Evaluation
We will train a Decision Tree Regressor and a Random Forest Regressor, then compare their performance using Mean Absolute Error and R-squared.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

#object of decision tree and random forest
dt_model=DecisionTreeRegressor()
rf_model=RandomForestRegressor(max_depth=20,
 min_samples_leaf=4,
 min_samples_split=2,
 n_estimators=100)

#train the model
```

```python
#we never rely on one scoring metrics for both regression and classification
```

```python

#train both the models of model set
dt_model.fit(x_train,y_train)
```

```python
rf_model.fit(x_train,y_train)
```

```python
#predicting target features
dt_model.predict(x_test)
```

```python
rf_model.predict(x_test)
```

```python
#check r2 score for both decision tree and random forest
#on train and testing set
#decision tree accuracy
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_train,dt_model.predict(x_train)))
print(f"mae for training set dt: {mean_absolute_error(y_train,dt_model.predict(x_train))}")
```

```python
print(r2_score(y_test,dt_model.predict(x_test)))
print(f"mae for testing set dt:{mean_absolute_error(y_test,dt_model.predict(x_test))}")
```

```python
#randomforst accuracy
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_train,rf_model.predict(x_train)))
print(f"mae for training set rf: {mean_absolute_error(y_train,rf_model.predict(x_train))}")
```

```python
print(r2_score(y_test,rf_model.predict(x_test)))
print(f"mae for testing set rf: {mean_absolute_error(y_test,rf_model.predict(x_test))}")
```

### 5. Hyperparameter Tuning
We will use `GridSearchCV` to find the best parameters for the Random Forest Regressor to improve its generalization on the test set.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'n_estimators':[100,200],        #num of decision tree models
    'max_depth':[10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
grid_search=GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    scoring='neg_mean_absolute_error',
    cv=3,
    n_jobs=-1,     #jitne bhi cores free hai they will perform parameter tunning
    verbose=1       #display the excution

)
```

```python
#fit the model for training set
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_rf_medels=grid_search.best_estimator_

#the best model we are going to pick is best_rf_model which have tunneed parameters
```

```python
tuned_predictions=best_rf_medels.predict(x_test)
```

```python
r2_score(y_test,tuned_predictions)
```

```python
mean_absolute_error(y_test,tuned_predictions)
```
